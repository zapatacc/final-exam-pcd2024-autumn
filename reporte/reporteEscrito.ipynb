### 2. Anlisis Exploratorio de los Datos (EDA)

Para explorar y preparar los datos, se cre un *notebook* dedicado, disponible en `notebooks/eda.ipynb`. Este contiene todo el desarrollo del anlisis exploratorio de los datos. A continuacin, se destacan los puntos principales:

- **Limpieza de los Datos**:  
  Los datos presentaban inconsistencias y requeran un tratamiento especfico. Algunos aspectos clave de este proceso fueron:
  - **Normalizacin**: Los datos estaban almacenados en formato JSON, por lo que se normalizaron para convertirlos en un formato tabular (*dataframe*).
  - **Filtrado de columnas relevantes**: Se seleccionaron nicamente las columnas necesarias para el anlisis:  
    - `"_source.complaint_what_happened"`  
    - `"_source.product"`  
    - `"_source.sub_product"`  
  - **Renombrado de columnas**: Las columnas se renombraron para mejorar su legibilidad y facilidad de uso en los siguientes pasos.

- **Ingeniera de Caractersticas**:  
  Se llev a cabo una pequea etapa de ingeniera de caractersticas para ajustar los datos a los requisitos del modelo. El resultado final de este proceso dej los datos en un formato limpio y estructurado, adecuado para su uso en las siguientes etapas del proyecto.

- **Tratado de Nulos**:  
  Durante la limpieza de los datos, se identific un caso particular con valores nulos. Aunque inicialmente pareca que no haba valores faltantes evidentes (`NaN`), se descubri que ciertos campos contenan un valor vaco (`""`), lo cual representa un carcter nulo (`'\0'`). 

  Para manejar este problema:
  - Se convirtieron estos valores vacos explcitos a `NaN` para facilitar su identificacin.
  - Posteriormente, se realiz un filtrado para eliminar estos registros nulos y garantizar la calidad de los datos.  

Este enfoque sistemtico asegur que los datos estuvieran preparados para entrenar modelos precisos y confiables, adems de proporcionar informacin valiosa para el desarrollo del resto del proyecto.
